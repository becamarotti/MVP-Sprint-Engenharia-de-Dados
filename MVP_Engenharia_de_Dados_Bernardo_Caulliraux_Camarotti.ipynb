## 2. Catálogo dos Dados



Após a conclusão da fase de 'busca e coleta de dados', entramos na etapa de catalogação. Neste estágio, realizamos uma investigação mais aprofundada para construir um catálogo mais completo e compreender melhor os dados disponíveis.

Os dados da base fictícia BaseFuncionarios que utilizaremos é estruturado em 25 colunas, sendo cada uma delas:

- **ID RH ->** Matrícula do colaborador
- **Nome Completo ->** Nome completo de cada colaborador
- **Estado Civil ->** Temos os estados C (casado) e S (solteiro)
- **Sexo ->** Temos M (masculino) e F (feminino)
- **Data Nascimento ->** Data de nascimento do colaborador
- **Endereço ->** Endereço da residência do colaborador
- **Data de Contratação ->** Data de admissão do colaborador
- **Data de Demissão ->** Caso o colaborador esteja ativo, a célula estará em branco, caso esteja demitido, a célula estará com a data de demissão
- **Salário Base ->** Salário mensal do colaborador antes dos impostos
- **Impostos ->** Valor do imposto sobre o salário
- **Benefícios ->** Valor mensal pago em benefícios ao colaborador
- **VT ->** Valor pago ao colaborador em vale transporte mensalmente
- **VR ->** Valor pago em vale refeição ao colaborador mensalmente
- **Cargo ->** Código do nível e área do colaborador
- **Nível ->** Qual nível do cargo do colaborador (Estagiário, Analista, Coordenador, Gerente e Diretor)
- **Área ->** Área de atuação do colaborador
- **Férias Acumuladas** -> Férias acumuladas no ano do colaborador (dias mínimo 0 e dias máximo 60)
- **Horas Extras ->** Quantidade de horas extras acumuladas no ano
- **Imagem ->** Imagem dos colaboradores
- **Avaliação ->** É a média das 4 notas referente ao Trabalho, Liderança, Comunicação, Iniciativa e Organização
- **Trabalho ->** Nota referente ao trabalho do colaborador (nota mínima 0 e nota máxima 10)
- **Liderança ->** Nota referente a liderança do colaborador (nota mínima 0 e nota máxima 10)
- **Comunicação ->** Nota referente a comunicação do colaborador (nota mínima 0 e nota máxima 10)
- **Iniciativa ->** Nota referente a iniciativa do colaborador (nota mínima 0 e nota máxima 10)
- **Organização ->** Nota referente a organização do colaborador (nota mínima 0 e nota máxima 10)

A base calendário está estruturada em 7 colunas. A base vai do dia 01 de janeiro de 2000 até o dia 31 de dezembro de 2018. Ela está estruturada da seguinte forma:

- **Data ->** Todas as datas de 01/01/2000 até 31/12/2018
- **Ano ->** Auxiliar da coluna **Data**. Apenas os anos de 2000 até 2018
- **Mês ->** Auxiliar da coluna **Data**. Os meses em formato de número. De 1 até 12.
- **Dia ->** Auxiliar da coluna **Data**. Apenas os dias.
- **Nome do Mês ->** Auxiliar da coluna **Data**. Os meses por extenso.
- **Início do Mês ->** Auxiliar da coluna **Data**. A data do primeiro dia do mês.
- **Mês/Ano ->** Auxiliar da coluna **Data**. As 3 primeiras letras do mês e o ano.

O Azure Data Factory é a ferramenta de ETL desenvolvida pela Microsoft para Extrair, Transformar e Carregar na nuvem. Ela permite a criação de fluxo de dados e pipelines para o carragemento dos dados no banco de dados.

Inicialmente, coletamos os dados armazenados no Data Lake. Em seguida, utilizamos o Data Factory como uma etapa intermediária, onde empregamos a ferramenta conhecida como DataFlow. O DataFlow é reconhecido por sua capacidade de depurar e preparar os dados, tornando-os mais acessíveis para a criação de análises.

O processo de DataFlow envolve a coleta de arquivos do Data Lake, realização dos ajustes necessários para a limpeza dos dados e, posteriormente, com o auxílio de um pipeline, executamos o fluxo de dados já depurados no Data Warehouse.
